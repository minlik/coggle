```python
# 步骤1：创建如下数据

data1 = pd.DataFrame({"ID":range(101, 106),          
                      "x1":range(1, 6),
                      "x2":["a", "b", "c", "d", "e"],
                      "x3":range(16, 11, - 1)})

data2 = pd.DataFrame({"ID":range(104, 108),        
                      "y1":["x", "y", "x", "y"],
                      "y2":range(8, 1, - 2)})

data3 = pd.DataFrame({"ID":range(102, 110),     
                      "z1":range(10, 18),
                      "z2":["z", "b", "z", "z", "d", "z", "d", "a"],
                      "z3":range(18, 10, - 1)})
```

```python
# 步骤3：Merge data1 and data2 Using Outer Join
data4 = pd.merge(data1, data2, on='ID', how='outer')

# 步骤4：Merge data1 and data2 Using Left Join
data5 = pd.merge(data1, data2, on='ID', how='left')

# 步骤5：Merge data1 and data2 Using Right Join
data6 = pd.merge(data1, data2, on='ID', how='right')

# 步骤6：Merge data1 ,data2 and data3 Using Outer Join
data7 = pd.merge(pd.merge(data1, data2, on='ID', how='outer'), data3, on='ID', how='outer')

# 步骤7：Merge data1 and data2 based on Index
data8 = data1.join(data2, lsuffix='_x', rsuffix='_y')
```
